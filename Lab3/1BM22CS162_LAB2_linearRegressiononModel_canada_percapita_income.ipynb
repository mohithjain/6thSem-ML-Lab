```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Load the data
data = pd.read_csv('/content/canada_per_capita_income.csv')
print("Data Preprocessing")
print(data.isnull().sum())
# Prepare the data
X = data['year'].values.reshape(-1, 1)
y = data['per capita income (US$)'].values

# Create and train the linear regression model
model = LinearRegression()
model.fit(X, y)

# Predict the per capita income for the year 2020
year_2020 = np.array([[2020]])
predicted_income = model.predict(year_2020)

print(f"Predicted per capita income in 2020: ${predicted_income[0]:.2f}")

# Plot the data and the regression line
plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X, model.predict(X), color='red', label='Regression Line')
plt.scatter(year_2020, predicted_income, color='green', label='Predicted 2020')
plt.xlabel('Year')
plt.ylabel('Per Capita Income (US$)')
plt.title('Canada Per Capita Income Over Years')
plt.legend()
plt.show()
```
